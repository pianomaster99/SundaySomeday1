```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
#Current path of this file
aiPath = Path().cwd()

#Current path of backend folder
backendPath = aiPath.parent

#Path of the chineseMNIST dataset
csv_path = backendPath / "data" / "chineseMNIST.csv"
```

```python
#Reading the dataset
df = pd.read_csv(csv_path)
print(df.head())
```

```python
#Get all unique values of the dataset
print(df["character"].unique())
"""
Labels
['九' '十' '百' '千' '万' '亿' '零' '一' '二' '三' '四' '五' '六' '七' '八']

Translated
['nine', 'ten', 'hundred', 'thousand', 'ten thousand', 'hundred million', 
'zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight']
"""

print(len(df["character"].unique()))
```

```python
firstRow = df.iloc[0] #Get first element of the dataframe
firstRowPixels = firstRow[:-2] #Only keep the pixels, ignore the label and chinese character
firstRowArray = firstRowPixels.to_numpy().astype(np.float32).reshape(64,64) #Convert the dataframe Series into numpy array and reshape into (64, 64) matrix
plt.imshow(firstRowArray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
def displayNImages(n: int, randomDisplay=False):
    
    _, ax = plt.subplots(n)
    if n == 1:
        ax = [ax]
    for i, ax in enumerate(ax):
        if randomDisplay:
            row = df.iloc[random.randint(1,len(df))]
        else:
            row = df.iloc[i]
        pixels = row[:-2].to_numpy().astype(np.float32)
        image = pixels.reshape(64, 64)
        ax.imshow(image)
        ax.axis('off')
        ax.set_title(f"Label: {row.iloc[-2]}, Chinese Character: {row.iloc[-1]}", fontname='STHeiti')
    plt.show()
```

```python
displayNImages(2, True)
```
